```python
import random
dataset = []
X = 0
Y = 0

names = np.array(data_filter["polish_name"])
emissions = np.array(data_filter["Total_emissions"])

X = []
Y = []
indexes = list(range(len(emissions)))
print(indexes)

for i in indexes:
    X.append(names[i])
    Y.append(emissions[i])
    
for i in range(100):
    random.shuffle(indexes)
    current_name = names[indexes[0]] + ", " + names[indexes[1]]
    print(names[indexes[0]] + ", " + names[indexes[1]])
    print(emissions[indexes[0]], emissions[indexes[1]], (emissions[indexes[0]] + emissions[indexes[1]])/2)
    X.append(current_name)
    current_emission = (emissions[indexes[0]] + emissions[indexes[1]])/2
    Y.append(current_emission)

for i in range(1000):
    random.shuffle(indexes)
    current_name = names[indexes[0]] + ", " + names[indexes[1]] + ", " + names[indexes[2]]
    X.append(current_name)
    current_emission = (emissions[indexes[0]] + emissions[indexes[1]] + emissions[indexes[2]])/3
    print(current_name)
    print(emissions[indexes[0]], emissions[indexes[1]], emissions[indexes[1]], current_emission)
    Y.append(current_emission)
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim

sentences = ["Piwo to moje paliwo", "Uwielbiam zajadać się słodkościami."]
model = SentenceTransformer("sdadas/st-polish-paraphrase-from-mpnet")
results = model.encode(sentences, convert_to_tensor=True, show_progress_bar=False)
print(cos_sim(results[0], results[1]))
print(results[0])
# tensor([[0.9794]], device='cuda:0')
```

```python
X = model.encode(X, convert_to_tensor=True, show_progress_bar=True)
print(X.shape)
X = X.cpu().numpy()
```

```python
Y = np.array(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn import svm
from sklearn.svm import SVR

regr = svm.SVR(C=10e10)
regr.fit(X_train, Y_train)

score = regr.score(X_test, Y_test)
Y_pred = regr.predict(X_test[:1])
```

```python
print(score, Y_pred)
```
